```python
crime_summary=df.groupby(['AREA NAME', 'Crm Cd Desc']).size().reset_index(name='count')
```

```python
# Her bölge için en çok görülen suçu seç
```

```python
top_crimes_by_area = crime_summary.sort_values('count', ascending=False).groupby('AREA NAME').first().reset_index()
```

```python
plt.figure(figsize=(16,10))
plt.barh(top_crimes_by_area['AREA NAME'], top_crimes_by_area['count'], edgecolor='black')
plt.title('Top crime by Area')
plt.tight_layout()
# Her cubugun yanina suc adini yaz 
for i, (area, crime, count) in enumerate(zip(top_crimes_by_area['AREA NAME'], 
                                             top_crimes_by_area['Crm Cd Desc'], 
                                             top_crimes_by_area['count'])): 
    plt.text(count+10, i, crime, va='center', fontsize=7)
    
```

```python
crime_count_per_date=df.groupby(['DATE OCC','Crm Cd Desc']).size().reset_index(name='average_daily_count')
```
